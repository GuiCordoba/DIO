## Analise de Dados com Python e Pandas

```python
# 1- Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn') 
```

```python
# 3 - Criando o DataFrame
df = pd.read_excel('AdventureWorks.xlsx') 
```

```python
# 5 - Verificando os tipos de dados
df.dtypes
```

# Verificando qual foi a receita e o lucro  de cada ano 

```python
# 5 - Receita total de cada ano
df.groupby(df['Data Venda'].dt.year)['Valor Venda'].sum().plot.bar(title='Receita x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita')
plt.xticks(rotation='horizontal');
```

```python
# Valores de receita total por ano
df_receita_por_ano = df.groupby(df['Data Venda'].dt.year)['Valor Venda'].sum() 
round(df_receita_por_ano)
```

```python
# 6 - Encontrando o lucro
# 6.1 - Primeiro vamos encontar o custo total 
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head()
```

```python
# 6.2 - Encontrando o Lucro
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head()
```

```python
# 6.3 - Grafico Lucro x Ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
# Valores de lucro por ano
df_lucro_por_ano = df.groupby(df['Data Venda'].dt.year)['Lucro'].sum() 
round(df_lucro_por_ano)
```

# Verificando os produtos mais vendidos

```python
# 7 - Produtos mais vendidos 
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```
